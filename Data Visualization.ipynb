```python
# Q.4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
# variable across different categories.
# Ans -- 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Step 1: Generate a dataset
np.random.seed(0)  # For reproducibility
categories = ['A', 'B', 'C', 'D', 'E']
data = {
    'Category': np.repeat(categories, 100),
    'Value': np.concatenate([
        np.random.normal(loc=i * 10, scale=5, size=100) for i in range(len(categories))
    ])
}

# Create a DataFrame
df = pd.DataFrame(data)

# Step 2: Create and display a box plot
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Box Plot of Numerical Values Across Categories')
plt.grid(True, linestyle='--', alpha=0.7)

# Step 3: Create and display a violin plot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
sns.violinplot(x='Category', y='Value', data=df, palette='Set2')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Violin Plot of Numerical Values Across Categories')
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()

```

```python
# Q.5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
# Ans --
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Step 1: Generate a synthetic dataset with correlated features
np.random.seed(0)  # For reproducibility

# Number of samples
n_samples = 100

# Generate random features
X1 = np.random.normal(loc=0, scale=1, size=n_samples)
X2 = 0.8 * X1 + np.random.normal(loc=0, scale=0.5, size=n_samples)  # X2 is correlated with X1
X3 = 0.5 * X1 + 0.5 * X2 + np.random.normal(loc=0, scale=0.5, size=n_samples)  # X3 is correlated with X1 and X2

# Create a DataFrame
df = pd.DataFrame({
    'Feature 1': X1,
    'Feature 2': X2,
    'Feature 3': X3
})

# Step 2: Compute the correlation matrix
correlation_matrix = df.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', linewidths=0.5)

# Customize the plot
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#Q1 -- Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
# Ans -- 
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Step 1: Generate the data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Sine of x values

# Step 2: Create the Bokeh plot
output_notebook()  # For displaying the plot inline in a Jupyter notebook

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

p.line(x, y, legend_label="Sine Wave", line_width=2, color="blue")

# Show the plot
show(p)
from bokeh.io import output_file

# Save the plot to an HTML file
output_file('sine_wave.html')

# Show the plot
show(p)
```

```python
# Q3 -- Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]
import bokeh.plotting as bp
from bokeh.embed import components
from bokeh.models import ColumnDataSource

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = bp.figure(x_range=fruits, height=350, title="Fruit Counts")
p.vbar(x='fruits', top='counts', width=0.9, color="firebrick", source=source, legend_field="fruits")

show(p)
```

```python
# Q4. --  Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)
# Ans -- 
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Category10

# Enable Bokeh output to notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Random Data", 
           x_axis_label='Value', 
           y_axis_label='Frequency',
           tools="save,hover",
           tooltips=[("Value", "@x"), ("Frequency", "@top")])

# Add histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color=Category10[10][0], 
       line_color='black')

# Display the plot
show(p)
```
